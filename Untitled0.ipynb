# Algoritmo KNN (k-nearest-neighbors)

```python
from sklearn.neighbors import KNeighborsClassifier

# 1. Dados mais encorpados (vários exemplos de cada)
# [Comprimento, Largura]
X = [
    [1.4, 0.2], [1.3, 0.2], [1.5, 0.3],  # Setosas (Pequenas)
    [4.5, 1.3], [4.0, 1.5], [4.7, 1.4],  # Versicolors (Médias)
    [6.1, 2.3], [5.8, 1.9], [6.5, 2.1]   # Virginicas (Grandes)
]
y = ['Setosa']*3 + ['Versicolor']*3 + ['Virginica']*3

# 2. Agora o K=3 faz sentido!
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# 3. Teste com um valor bem alto (deve ser Virginica)
previsao_grande = knn.predict([[6.8, 2.5]])
# Teste com um valor bem baixo (deve ser Setosa)
previsao_pequena = knn.predict([[1.1, 0.1]])

print(f"Flor grande: {previsao_grande[0]}")
print(f"Flor pequena: {previsao_pequena[0]}")
```
